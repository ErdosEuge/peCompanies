```python
import pandas as pd
from sklearn.preprocessing import StandardScaler
```

```python
df = pd.read_csv("/workspaces/peCompanies/companies.csv")
df.head()
```

```python
df_proc = df.copy()
# Salva l'ultima colonna (classe) in una variabile separata
classe = df['Company']

# Rimuovi l'ultima colonna dal dataset
df = df.drop('Company', axis=1)

# Seleziona le colonne che desideri codificare con one-hot encoding
colonnes_one_hot = ['Sector', 'Industry']
encoded_dataset = pd.get_dummies(df, columns=colonnes_one_hot)

# Esegui la standardizzazione gaussiana sulle colonne
scaler = StandardScaler()
scaled_dataset = scaler.fit_transform(encoded_dataset)

# Converti il dataset standardizzato in un nuovo DataFrame
scaled_df = pd.DataFrame(scaled_dataset, columns=encoded_dataset.columns)

# Ora puoi utilizzare scaled_df per ulteriori analisi o salvataggio
# scaled_df['Company'] = classe

scaled_df
```

```python
scaled_df.to_csv('./df_total.tsv', sep='\t', index=False, header=False)
df_proc.loc[:, ['Company', 'Industry', 'Sector']].to_csv('./comp_ind_sec.tsv', sep='\t', index=False)
```
